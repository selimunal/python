```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('../TensorFlow_FILES/DATA/fake_reg.csv')
df.head()
```

```python
df.shape
```

```python
sns.pairplot(df)
```

```python
from  sklearn.model_selection import train_test_split
```

```python
X=df['price'].drop
X
```

```python
X=df[['feature1','feature2']].values
```

```python
X.shape
```

```python
y=df['price'].values
```

```python
y.shape
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#help(MinMaxScaler)
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train.max()
```

```python
X_train.min()
```

```python
X_test.mean()
```

```python
X_train.mean()
```

```python
import matplotlib.pyplot as plt
```

```python
X_train[:,0].shape
```

```python
plt.plot(X_train[:,0],y_train)
```

```python
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense
```

```python
#help(Dense)
```

```python
model = Sequential([Dense(4,activation='relu'),
                   Dense(2,activation='relu'),
                   Dense(1)])
```

```python
model=Sequential()

model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

model.add(Dense(1))#predict the price
#THIS IS SAME AS ABOVE MODEL = comment out to edit-delete layers

model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=250)
```

```python
loss_df=pd.DataFrame(model.history.history)
loss_df.head()
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test,y_test,verbose=0) #metric result
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions=model.predict(X_test)
```

```python
#test_predictions
```

```python
test_predictions=pd.Series(test_predictions.reshape(300,))
test_predictions #it is pandas series now
```

```python
pred_df=pd.DataFrame(y_test,columns=['Test True Y'])
pred_df.head()
```

```python
pred_df=pd.concat([pred_df,test_predictions],axis=1)
```

```python
pred_df.columns=['Test True Y','Model Predictions']
pred_df.head()
```

```python
sns.scatterplot(data=pred_df,x='Test True Y',y='Model Predictions')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```

```python
mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```

```python
#df.describe()
```

```python
root_mean_sq_error=mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions'])**0.5
```
