```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

#larger figure axes1
axes1.plot(x,y,"b")
axes1.set_xlabel("X_label_axes2")
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

#insert figure axes 2
axes2.plot(y,x,"r")
axes2.set_xlabel("X_label_axes2")
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 title')
```

```python
#use similar to plt.figure() except use tuple unpacking
#to grab fig and axes
fig,axes=plt.subplots()
axes.plot(x,y,"r")

axes.plot(x,y,'r')
axes.set_xlabel("x")
axes.set_ylabel('y')
axes.set_title('title')
```

```python
#empty canvas of 1 by subplots
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
axes
```

```python
for ax in axes:
    ax.plot(x,y,"b")
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
fig
```

```python
fig, axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('first plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')

plt.tight_layout()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,label='x,y')
ax.plot(x,x**4,label='x uzeri 4')
ax.plot(x,x**3,label='x cubed')
ax.plot(y,x,label='y,x')

ax.legend(loc=10)
```

# plot colors

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,label='x,y',color='red',linewidth=3,alpha=0.6,linestyle='--')
ax.plot(x,x**4,label='x uzeri 4',color="purple",linewidth=6,alpha=0.5,linestyle='-.')
ax.plot(x,x**3,label='x cubed',lw=2,ls=':')
ax.plot(y,x,label='y,x',lw=15,ls='--',marker='o')

ax.legend(loc=10)
```

# y and x limit

```python
fig=plt.figure(figsize=(5,2))

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,lw=2,markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
