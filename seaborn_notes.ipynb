```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head(10)
```

```python
sns.set()
sns.distplot(tips["total_bill"],kde=False)
```

```python
sns.displot(tips["total_bill"],bins=100)
```

```python
sns.displot(tips["total_bill"],bins=30)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
#import matplotlib as mpl

#mpl.rcParams['patch.force_edgecolor'] = True
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
sns.pairplot(tips)#quick visualize
```

```python
#import matplotlib as mpl

#mpl.rcParams['patch.force_edgecolor'] = True

sns.pairplot(tips,hue='sex',palette='coolwarm',diag_kind="hist")
```

```python
sns.rugplot(tips["total_bill"],height=0.1,expand_margins=False)
```

```python
sns.distplot(tips["total_bill"])
```

```python
import matplotlib.pyplot as plt
sns.kdeplot(tips["total_bill"],label="total bill")
plt.legend()
```
