```python
tips.info()
```

```python
flights.info()
```

```python
tc= tips.corr() #u need a matrix form
```

```python
sns.heatmap(tc) #matrix form is mapped
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights #you need this in matrix form
```

```python
flights.pivot_table(index='month',columns='year',values='passengers') 
#now looks matrix
```

```python
fp= flights.pivot_table(index='month',columns='year',values='passengers') 
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=3)
```

```python
sns.heatmap(fp,cmap='magma',linecolor='black',linewidths=3)
```

# clustermap

```python
sns.clustermap(fp)
#similar puts together, similar to each other put together
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

# GRIDS

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.set_theme(style="whitegrid")

sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g= sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
 g= sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# Regression Plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
#scatter_kws changes marker size
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
#seperate by columns
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',
          aspect=0.6,height=3)
```
